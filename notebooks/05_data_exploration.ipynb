```python
import tensorflow as tf
import glob
from torchvision import transforms
from pathlib import Path
from PIL import Image
import matplotlib.pyplot as plt
import numpy as np

def make_x_train(glob_path):
    x_train = []
    
    for i in glob.glob(glob_path):
        image = Image.open(str(i))
        image = np.asarray(image)
        x_train.append(image)
    
    return np.array(x_train)

def stack_images(x_train, figure_title):
    image_stacks = np.zeros(shape=(len(x_train), 224, 224, 3), dtype=np.uint8)

    for i in range(len(x_train)):
        # open image file and store in variable `im`, then
        image_stacks[i] = x_train[i]
        
    fig = plt.figure(figsize=(10, 4.9))
    fig.tight_layout()
    fig.suptitle(figure_title, fontweight='bold')

    image_median = np.mean(image_stacks, axis=0).astype(np.uint8)
    fig.add_subplot(1, 2, 1)
    img = Image.fromarray(image_median)
    plt.imshow(img)
    plt.axis('off')
    plt.title('Mean')
    
    
    image_median = np.median(image_stacks, axis=0).astype(np.uint8)
    fig.add_subplot(1, 2, 2)
    img = Image.fromarray(image_median)
    plt.imshow(img)
    plt.axis('off')
    plt.title('Median')
```

# Christmas

```python
x_train = make_x_train("../data/3_Train_Test_Folder/train/christmas/*")
stack_images(x_train, 'Christmas')
```

# Normal

```python
x_train = make_x_train("../data/3_Train_Test_Folder/train/normal/*")
stack_images(x_train, 'Normal')
```
